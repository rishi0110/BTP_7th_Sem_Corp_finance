```python
import yfinance as yf
import os

# List of tickers
tickers = [
    "ADANIENT.NS", "WIPRO.NS", "LT.NS", "APOLLOHOSP.NS", "TATACONSUM.NS",
    "BAJAJFINSV.NS", "BHARTIARTL.NS", "TRENT.NS", "MARUTI.NS", "ITC.NS",
    "SHRIRAMFIN.NS", "TCS.NS", "BAJAJ-AUTO.NS", "COALINDIA.NS", "BAJFINANCE.NS",
    "HDFCLIFE.NS", "RELIANCE.NS", "CIPLA.NS", "BRITANNIA.NS", "HEROMOTOCO.NS",
    "INDUSINDBK.NS", "HINDALCO.NS", "ULTRACEMCO.NS", "KOTAKBANK.NS", "NTPC.NS",
    "TATASTEEL.NS", "TITAN.NS", "NESTLEIND.NS", "ONGC.NS"
]

# Create folders for saving CSV files
os.makedirs("balance_sheet", exist_ok=True)
os.makedirs("cash_flow", exist_ok=True)
os.makedirs("income_statement", exist_ok=True)

# Loop through each ticker
for ticker in tickers:
    # Get ticker data
    rel = yf.Ticker(ticker)
    
    # Fetch balance sheet, cash flow, and income statement data
    balance_sheet = rel.balance_sheet
    cash_flow = rel.cashflow
    income_statement = rel.financials  # Use .financials for the income statement

    # Save balance sheet data
    if not balance_sheet.empty:
        balance_sheet_filename = f"balance_sheet/{ticker}_balance_sheet.csv"
        balance_sheet.to_csv(balance_sheet_filename)
        print(f"Saved {balance_sheet_filename}")

    # Save cash flow data
    if not cash_flow.empty:
        cash_flow_filename = f"cash_flow/{ticker}_cash_flow.csv"
        cash_flow.to_csv(cash_flow_filename)
        print(f"Saved {cash_flow_filename}")

    # Save income statement data
    if not income_statement.empty:
        income_statement_filename = f"income_statement/{ticker}_income_statement.csv"
        income_statement.to_csv(income_statement_filename)
        print(f"Saved {income_statement_filename}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import random

# Set the path to your CSV files
path = r'E:\BTP\output\output\*.csv'

# Get a list of all CSV files in the directory
all_files = glob.glob(path)

# Randomly select 5 CSV files
selected_files = random.sample(all_files, 5)

# Set the style for Seaborn
sns.set(style="whitegrid")

# Loop through the selected files and create plots
for file in selected_files:
    # Load the data
    df = pd.read_csv(file)
    
    # Set 'Unnamed: 0' as the index and rename it to 'year'
    df.set_index('Unnamed: 0', inplace=True)
    df.index.name = 'year'
    
    # Transpose the DataFrame
    df_transposed = df.T
    
    # Extract the first 4 characters from the year index, convert to int, and subtract 1
    new_years = df_transposed.index.str[:4].astype(int) - 1
    
    # Update the index with the modified years
    df_transposed.index = new_years
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    
    # Check if 'TLR' is in the DataFrame
    if 'TLR' in df_transposed.columns:
        sns.lineplot(data=df_transposed, x=df_transposed.index, y='LTLR', marker='o')
        
        # Customize the plot
        plt.title(f"Plot of 'LTLR' vs. Years from {file[28:]}", fontsize=16)
        plt.xlabel("Year", fontsize=14)
        plt.ylabel("TLR", fontsize=14)
        plt.xticks(df_transposed.index)  # Set x-ticks to display each year
        plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
        plt.grid(True)

        # Show the plot
        plt.tight_layout()
        plt.show()
    else:
        print(f"'TLR' column not found in {file}.")
```
